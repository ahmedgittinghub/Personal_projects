```python

def returns_calculator(list_name, close_data_frame ,open_data_frame, new_list):
    dict_1 = {'ticker_name': '' , 'stock_returns' : '', 'stock_%returns': '' }

    current_date = datetime.now()
    end_date = current_date - timedelta(days=1)
    start_date = current_date - timedelta(days=7)

    # Format the start date as a string
    start_date_str = start_date.strftime("%Y-%m-%d")
    current_date_str = current_date.strftime("%Y-%m-%d")
    end_date_str = end_date.strftime("%Y-%m-%d")

    for ticker in list_name:
        dict_1['ticker_name'] = ticker
        # calculating  & rounding up returns 
        dict_1['stock_returns'] = (close_data_frame.loc[end_date_str, ticker] - open_df.loc[start_date_str, ticker]).round(2)
        # Converting to a string
        dict_1['stock_returns'] = '$' + str(dict_1['stock_returns'])

        # calculating percentage returns 
        dict_1['stock_%returns'] = ((close_data_frame.loc[end_date_str, ticker] - open_data_frame.loc[start_date_str, ticker]) * 100/(open_data_frame.loc[start_date_str,ticker])).round(2)
        dict_1['stock_%returns'] = '%' + str(dict_1['stock_%returns'])
        new_list.append(dict_1)
        dict_1 = {'ticker_name': '' , 'stock_returns' : '', 'stock_%returns': '' }
    
    return new_list
    
```

```python
# Now I can turn this loop into a function for a much cleaner code 

def stock_volatility_calculator(new_list, list_of_dates, list_of_tickers, close_data, open_data,):
    
    day_1 = list_of_dates[0]
    day_2 = list_of_dates[1]
    day_3 = list_of_dates[2]
    day_4 = list_of_dates[3]
    day_5 = list_of_dates[4]

    dict_2 = {'Stock_Ticker':'', 'Stock_Volatility':''}

    for ticker in list_of_tickers:
        day_1_returns = ((close_data.loc[day_1, tickers] - open_data.loc[day_1, tickers]) * 100/(open_data.loc[day_1,tickers]))
        day_2_returns = ((close_data.loc[day_2, tickers] - open_data.loc[day_2, tickers]) * 100/(open_data.loc[day_2,tickers]))
        day_3_returns = ((close_data.loc[day_3, tickers] - open_data.loc[day_3, tickers]) * 100/(open_data.loc[day_3,tickers]))
        day_4_returns = ((close_data.loc[day_4, tickers] - open_data.loc[day_4, tickers]) * 100/(open_data.loc[day_4,tickers]))
        day_5_returns = ((close_data.loc[day_5, tickers] - open_data.loc[day_5, tickers]) * 100/(open_data.loc[day_5,tickers]))

        mean_return = (day_1_returns + day_2_returns + day_3_returns + day_4_returns + day_5_returns) / 5

        variance = ((day_1_returns - mean_return)**2 + (day_2_returns - mean_return)**2 + (day_3_returns - mean_return)**2 + (day_4_returns - mean_return)**2 + (day_5_returns - mean_return)**2) / 4

        standard_devation = (variance**0.5).round(2)

        dict_2['Stock_Ticker'] = ticker

        dict_2['Stock_Volatility'] = standard_devation

        new_list.append(dict_2)

        dict_2 = {'Stock_Ticker':'', 'Stock_Volatility':''}
    
    return new_list


    
    
```

```python
def ticker_extractor(list_name):
    
    resp = requests.get('http://en.wikipedia.org/wiki/List_of_S%26P_500_companies')
    soup = bs.BeautifulSoup(resp.text, 'lxml')
    table = soup.find('table', {'class': 'wikitable sortable'})

    

    for row in table.findAll('tr')[1:]:
        ticker = row.findAll('td')[0].text
        list_name.append(ticker.replace('\n', ''))
        
    

    return list_name
    
```
